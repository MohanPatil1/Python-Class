```python
#Matplotlib
from matplotlib import pyplot as plt
```

```python
cars_data.corr()
```

```python
plt.scatter(cars_data['Age'],cars_data['Price'],color='green')
plt.title('Scatter plot of Price vs Age')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

```python
plt.hist(cars_data['KM'],color='green',edgecolor='white')
plt.title('Histogram')
plt.xlabel('KM')
plt.ylabel('Frequency')
plt.figure(figsize=(10,5))
plt.savefig("Bar jpeg")
plt.show()
```
